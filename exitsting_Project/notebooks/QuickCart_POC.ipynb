```python
def get_priority(row):

    rating = row["rating"]

    sentiment = row["sentiment"]

    if sentiment == "Negative":

        if pd.notna(rating):

            if rating <= 2:
                return "High"

            elif rating <= 3:
                return "Medium"

        return "Medium"

    return "Low"

clean_df["priority"] = (
    clean_df.apply(
        get_priority,
        axis=1
    )
)
```

```python
positive_words = [
    "great", "excellent", "good", "amazing",
    "love", "fantastic", "smooth", "fast",
    "helpful", "easy", "perfect"
]

negative_words = [
    "bad",
    "worst",
    "late",
    "delay",
    "crash",
    "error",
    "refund",
    "charged",
    "missing",
    "broken",
    "rude",
    "issue",
    "problem",
    "failed",
    "cancelled",
    "cannot",
    "cant",
    "did not",
    "not working",
    "stuck",
    "greyed out",
    "greyed out",
    "slow",
    "lost",
    "wrong",
    "never arrived",
    "not delivered"
]

def get_sentiment(text):

    text = str(text).lower()

    pos = sum(word in text for word in positive_words)
    neg = sum(word in text for word in negative_words)

    if pos > neg:
        return "Positive"

    elif neg > pos:
        return "Negative"

    return "Neutral"

clean_df["sentiment"] = (
    clean_df["feedback_clean"]
    .apply(get_sentiment)
)
```

```python
def classify_category(text):

    text = str(text).lower()

    if any(word in text for word in [
        "refund","charged","payment",
        "coupon","billing","subscription"
    ]):
        return "Billing"

    elif any(word in text for word in [
        "crash","bug","login","error",
        "freeze","app"
    ]):
        return "App Bug"

    elif any(word in text for word in [
        "delivery","driver","package",
        "late","missing item"
    ]):
        return "Delivery"

    elif any(word in text for word in [
        "support","agent","staff",
        "customer service"
    ]):
        return "Staff/Support"

    else:
        return "Other"

clean_df["category"] = clean_df["feedback_clean"].apply(classify_category)
```
